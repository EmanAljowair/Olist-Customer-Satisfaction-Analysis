```python
import pandas as pd
df = pd.read_csv('olist_order_reviews_dataset.csv')
```

```python
#getting the 1 scored review for further analysis
bad_df = df[(df['review_score'] == 1)].copy()
```

```python
bad_df = bad_df.dropna(subset=['review_comment_message'])
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-xnli-multilingual-nli-2mil7")
```

```python
issue_labels = [
    "Product Quality Issue",
    "Delivery Delay Issue",
    "Seller Issues",
    "Wrong Item Received",
    "Incomplete Order",
    "Packaging Issue",
    "Refund/Return Problem",
    "Payment/Checkout Issue",
    "Customer Service Issue",
    "App/Website Problem",
    "No Issue / Positive"
]
```

```python
def classify_issues(text):
    result = pipe(text, candidate_labels=issue_labels)
    return pd.Series(result['labels'][0])
```

```python
bad_df.shape
```

```python
bad_df['issue_classification']  = bad_df['review_comment_message'].apply(classify_issues)
```

```python
bad_df.to_csv("bad_df.csv")
```
